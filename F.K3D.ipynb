```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import pandas as pd
import os
plt.style.context(['science', 'no-latex', 'cjk-tc-font'])
# Check we are in examples dir
current_dir = os.getcwd().lower()
if (current_dir.endswith('scienceplots')):
    os.chdir('./examples')
# Create 'figures' folder if it does not exist
if (not os.path.exists('./figures')):
    os.makedirs('figures')

def model(p ,data):
    return data[p]


pparam = dict(xlabel='Voltage (mV)', ylabel=r'Current ($\mu$A)')
data = pd.read_csv("ImageNet100B0Inc5.csv")
x = data['Number of Classes']
L = ["K3D",	"iCaRL",	"BiC",	"WA",	"LwF",	"EWC", 	"RMM",	"Coil",	"MEMO"]

with plt.style.context(['science', 'high-vis']):
    fig, ax = plt.subplots()
    for p in L:
        ax.plot(x, model(p,data) , label=p)
    ax.legend(title='Order')
    ax.autoscale(tight=True)
    ax.set(**pparam)
    fig.savefig('figures/fig01.jpg', dpi=300)
    plt.close()
```
